```python
import tkinter as tk
import requests
import urllib.request
from bs4 import BeautifulSoup

class App:  
    def __init__(self):
        data = urllib.request.urlopen("https://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt")
        soup = BeautifulSoup(data, "html.parser")
        a1 = soup.find_all("p", class_="listName")
        a2 = a1[0].find("a").text
        b1 = soup.find_all("p", class_="listArea")
        for l, n in zip(a1, b1):
            count = 0
            a = count,'ìœ„ ', l.text.strip(), n.text.strip()
            count+=1
        self.root=tk.Tk()
        self.vsb = tk.Scrollbar(orient="vertical", command=self.OnVsb)
        self.lb = tk.Listbox(self.root, yscrollcommand=self.vsb.set)
        self.vsb.pack(side="right",fill="y")
        self.lb.pack(side="left",fill="x", expand=True)
        self.lb.bind("<MouseWheel>", self.OnMouseWheel)
        for l, n in zip(a1, b1):
            a = count,'ìœ„ ', l.text.strip(), n.text.strip()
            count+=1
            self.lb.insert(END, a,'\n')
        self.root.mainloop()

    def OnVsb(self, *args):
        self.lb.yview(*args)

    def OnMouseWheel(self, event):
        self.lb.yview("scroll", event.delta,"units")
        return "break"

app=App()
```
